Load test/train data

```python
import numpy as np
# Ensure data_dir is set correctly
data_dir = '/content/gdrive/MyDrive/NLP Project/data/'

x_train_path = data_dir + "X_train_simplified.txt"
x_test_path = data_dir + "X_test_simplified.txt"
y_train_path = data_dir + "y_train_simplified.txt"
y_test_path = data_dir + "y_test_simplified.txt"

# Load X_train (sentences) as text
with open(x_train_path, 'r') as f:
    X_train = f.read().splitlines()

# Load X_test (sentences) as text
with open(x_test_path, 'r') as f:
    X_test = f.read().splitlines()

# Load y_train (labels) as text and convert to int
with open(y_train_path, 'r') as f:
    y_train = [line.strip() for line in f.read().splitlines()]

# Load y_test (labels) as text and convert to int
with open(y_test_path, 'r') as f:
    y_test = [line.strip() for line in f.read().splitlines()]

print(f"Loaded X_train samples: {len(X_train)}")
print(f"Loaded X_test samples: {len(X_test)}")
print(f"Loaded y_train samples: {len(y_train)}")
print(f"Loaded y_test samples: {len(y_test)}")
```

Create chart for the split data

```python
import pandas as pd
from collections import Counter

categories = ['A', 'B', 'C']

# Map numerical categories from y_train/y_test to A, B, C
category_map = {'1': 'A', '2': 'B', '3': 'C'}

# Calculate counts for y_train
raw_train_counts = Counter(y_train)
train_counts = {category_map.get(k, k): v for k, v in raw_train_counts.items()}
train_stats = {cat: train_counts.get(cat, 0) for cat in categories}
train_stats['Total'] = len(y_train)

# Calculate counts for y_test
raw_test_counts = Counter(y_test)
test_counts = {category_map.get(k, k): v for k, v in raw_test_counts.items()}
test_stats = {cat: test_counts.get(cat, 0) for cat in categories}
test_stats['Total'] = len(y_test)

# Hardcoded GITHUB and KAGGLE stats as provided by setup data
github_stats = {'A': 3131, 'B': 6238, 'C': 675, 'Total': 10044}
kaggle_stats = {'A': 4176, 'B': 0, 'C': 1602, 'Total': 5778}

# Create a dictionary to hold all data for the DataFrame with the desired order
data = {
    'Source': ['GITHUB', 'KAGGLE', 'TOTAL', 'TEST', 'TRAIN']
}

# Initialize category and total columns with placeholders
for cat in categories:
    data[cat] = [0] * len(data['Source'])
data['Total'] = [0] * len(data['Source'])

# Populate data for GITHUB
data['A'][0] = github_stats['A']
data['B'][0] = github_stats['B']
data['C'][0] = github_stats['C']
data['Total'][0] = github_stats['Total']

# Populate data for KAGGLE
data['A'][1] = kaggle_stats['A']
data['B'][1] = kaggle_stats['B']
data['C'][1] = kaggle_stats['C']
data['Total'][1] = kaggle_stats['Total']

# Populate data for TEST
data['A'][3] = test_stats['A']
data['B'][3] = test_stats['B']
data['C'][3] = test_stats['C']
data['Total'][3] = test_stats['Total']

# Populate data for TRAIN
data['A'][4] = train_stats['A']
data['B'][4] = train_stats['B']
data['C'][4] = train_stats['C']
data['Total'][4] = train_stats['Total']

# Calculate and populate data for the overall TOTAL row (sum of all distinct sources)
for cat in categories:
    data[cat][2] = data[cat][0] + data[cat][1] + data[cat][3] + data[cat][4] # GITHUB + KAGGLE + TEST + TRAIN
data['Total'][2] = data['Total'][0] + data['Total'][1] + data['Total'][3] + data['Total'][4] # Sum of all totals

# Create DataFrame
df_stats = pd.DataFrame(data)

# Function to apply styling for borders
def apply_border_style(row):
    styles = [''] * len(row)
    if row['Source'] == 'KAGGLE':
        # Add a thicker bottom border after KAGGLE
        styles = ['border-bottom: 2px solid black'] * len(row)
    elif row['Source'] == 'TOTAL':
        # Add a thicker bottom border after TOTAL
        styles = ['border-bottom: 2px solid black'] * len(row)
    return styles

# Apply the styling to the DataFrame and display
display(df_stats.style.apply(apply_border_style, axis=1))
```
