```python
from google.colab import drive
drive.mount('/content/gdrive')

data_dir = '/content/gdrive/MyDrive/NLP Project/data/'

x_train_path = data_dir + "X_train.txt"
x_test_path  = data_dir + "X_test.txt"
y_train_path = data_dir + "y_train.txt"
y_test_path  = data_dir + "y_test.txt"
```

```python
# Sentences
with open(x_train_path, 'r') as f:
    X_train = f.read().splitlines()

with open(x_test_path, 'r') as f:
    X_test = f.read().splitlines()

# Labels
with open(y_train_path, 'r') as f:
    y_train = f.read().splitlines()

with open(y_test_path, 'r') as f:
    y_test = f.read().splitlines()
```

```python
def simplify_label(label):
    return label[0]   # first character only

y_train_simplified = [simplify_label(label.strip()) for label in y_train]
y_test_simplified  = [simplify_label(label.strip()) for label in y_test]
```

```python
x_train_out = data_dir + "X_train_simplified.txt"
x_test_out  = data_dir + "X_test_simplified.txt"
y_train_out = data_dir + "y_train_simplified.txt"
y_test_out  = data_dir + "y_test_simplified.txt"
```

```python
with open(x_train_out, 'w') as f:
    f.write("\n".join(X_train))

with open(x_test_out, 'w') as f:
    f.write("\n".join(X_test))

with open(y_train_out, 'w') as f:
    f.write("\n".join(y_train_simplified))

with open(y_test_out, 'w') as f:
    f.write("\n".join(y_test_simplified))

print("Simplified dataset files created successfully.")
```
